# Fixed-Point Iteration Convergence 

### Step 1: Setup
Since \(g\) is smoothly differentiable, its derivative \(g'\) is also continuous.

At the fixed point \(p\), we know:
$$
g(p)=p,\qquad |g'(p)|<1.
$$

### Step 2: Uniform Bound Near \(p\)
Because \(|g'(p)|<1\) and \(g'\) is continuous, I can find a number \(k\) such that
$$
|g'(p)|<k<1.
$$

By continuity of \(g'\), we can pick a small interval around \(p\), say \((p-\delta,\,p+\delta)\), where
$$
|g'(x)|\le k,\qquad \text{for all } x\in(p-\delta,\,p+\delta).
$$

This means near the fixed point, the derivative always stays below 1 in size.


### Step 3: Contraction Property
Look at the iteration:
$$
p_n = g(p_{n-1}).
$$

By the **Mean Value Theorem**, for each step there is some point \(\xi_n\) between \(p_{n-1}\) and \(p\) such that
$$
|p_n-p|=|g(p_{n-1})-g(p)| = |g'(\xi_n)|\,|p_{n-1}-p|.
$$

Since \(|g'(\xi_n)|\le k<1\),
$$
|p_n-p| \le k\,|p_{n-1}-p|.
$$

This shows each step moves closer to \(p\).


### Step 4: Induction
Applying this repeatedly,
$$
|p_n-p| \le k^n\,|p_0-p|.
$$

Because \(0<k<1\),
$$
\lim_{n\to\infty} k^n = 0 \quad \Rightarrow \quad \lim_{n\to\infty} |p_n-p|=0.
$$

The sequence {pn} converges to the fixed point p as n goes to infinity.
