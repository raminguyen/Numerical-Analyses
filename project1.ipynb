```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.exp(x) - x**2 + 3*x - 2

def bisection(a, b, tol, max_it):
    i = 1
    fa = f(a)
    fb = f(b)
    if fa * fb > 0:
        return np.array([])
    iters = []

    while i <= max_it:
        p = (a + b) / 2.0
        iters.append(p)
        fp = f(p)

        if fp == 0 or (b - a)/2.0 < tol:
            return np.array(iters, dtype=float)
        
        i += 1

        if fa * fp > 0:

            a = p
            fa = fp

        else:
            b = p

    return np.array(iters, dtype=float)

a, b   = 0.0, 1.0
tol    = 1e-5
max_it = 100
N_PTS  = 400

pn = bisection(a, b, tol, max_it)

p = pn[-1]

print(p)
print(f(p))

x = np.linspace(a, b, N_PTS)
y = f(x)

plt.plot(x, y, color='b')

plt.hlines(0, a, b, color='k')

plt.plot(pn, np.zeros_like(pn), 'ro-', ms=4, lw=1)

plt.scatter([p], [0], color='gold', edgecolors='k', s=80)

plt.show()
```
